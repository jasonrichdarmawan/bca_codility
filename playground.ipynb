```python
# you can write to stdout for debugging purposes, e.g.
# print("this is a debug message")
def solution(A):
    # Implement your solution here
    minA = min(i for i in A if i > 0)
    maxA = max(A)
    if maxA < 0:
        return 1
    
    listA = [x for x in range(minA, maxA + 1)]
    for a in A:
        try:
            listA.remove(a)
        except ValueError:
            continue

    # [0, 1] -> 2
    if len(listA) == 0:
        return maxA + 1
    
    return min(listA)
    
solution([-3, 3, 6, 4, 1, 2])
```

```python
def solution(A):
    # Implement your solution here
    A.sort()
    
    count = 0
    total = 0
    for idx, x in enumerate(A):
        if x > 0:
            count += 1
            total += x
        if count == 3:
            break
        
    return total

solution([4,-2,0,5,-2,1,6])
solution([3,-2,0])
solution([-9,-4,-2,-6])
```
